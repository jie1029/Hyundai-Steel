```python
import pandas as pd
data = pd.read_csv('./plant2_train_linear.csv', parse_dates=["plant2_train.mea_ddhr"])

del data["Unnamed: 0"]
del data["Unnamed: 0.1"]

data.rename(columns={
    "plant2_train.mea_ddhr" : "date",
    "plant2_train.tem_in_loc1" : "tem_loc1",
    "plant2_train.hum_in_loc1" : "hum_loc1",
    "plant2_train.tem_coil_loc1" : "coil_loc1",
    "plant2_train.tem_in_loc2" : "tem_loc2",
    "plant2_train.hum_in_loc2" : "hum_loc2",
    "plant2_train.tem_coil_loc2" : "coil_loc2",
    "plant2_train.tem_in_loc3" : "tem_loc3",
    "plant2_train.hum_in_loc3" : "hum_loc3",
    "plant2_train.tem_coil_loc3" : "coil_loc3",
    "plant2_train.tem_out_loc1" : "tem_out",
    "plant2_train.hum_out_loc1" : "hum_out",
    "plant2_train.cond_loc1" : "cond_loc1",
    "plant2_train.cond_loc2" : "cond_loc2",
    "plant2_train.cond_loc3" : "cond_loc3"
}, inplace = True)


data["date"] = data["date"].apply(lambda t: t.strftime('%Y-%m-%d %H:%M'))
data["X24H_TMA"] = pd.to_datetime(data["date"]) + pd.DateOffset(days=1)
data["X48H_TMA"] = pd.to_datetime(data["date"]) + pd.DateOffset(days=2)
data["X24H_TMA"] = data["X24H_TMA"].apply(lambda t: t.strftime('%Y-%m-%d %H:%M'))
data["X48H_TMA"] = data["X48H_TMA"].apply(lambda t: t.strftime('%Y-%m-%d %H:%M'))

label_data1 = data[['date','cond_loc1','cond_loc2','cond_loc3']]
label_data2 = data[['date','cond_loc1','cond_loc2','cond_loc3']]

label_data1.rename(columns = {
    "date":"X24H_TMA",
    "cond_loc1":"X24H_cond_loc1",
    "cond_loc2":"X24H_cond_loc2",
    "cond_loc3":"X24H_cond_loc3"
}, inplace = True)


label_data2.rename(columns = {
    "date":"X48H_TMA",
    "cond_loc1":"X48H_cond_loc1",
    "cond_loc2":"X48H_cond_loc2",
    "cond_loc3":"X48H_cond_loc3"
}, inplace = True)

merge_data = pd.merge(data, label_data1, how='left', on='X24H_TMA')
result_data = pd.merge(merge_data, label_data2, how='left', on='X48H_TMA')

result_data = result_data[[
    'date',
 'tem_loc1',
 'hum_loc1',
 'coil_loc1',
 'tem_loc2',
 'hum_loc2',
 'coil_loc2',
 'tem_loc3',
 'hum_loc3',
 'coil_loc3',
 'tem_out',
 'hum_out',
 'cond_loc1',
 'cond_loc2',
 'cond_loc3',
 'X24H_TMA',
 'X24H_cond_loc1',
 'X24H_cond_loc2',
 'X24H_cond_loc3',
 'X48H_TMA',
 'X48H_cond_loc1',
 'X48H_cond_loc2',
 'X48H_cond_loc3'
]]
```

```python
result_data.to_csv("plant2_linear_test.csv", mode='w')
```
